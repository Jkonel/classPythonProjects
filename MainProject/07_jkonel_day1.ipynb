3. secs=12305 单位是秒

    请输出格式为 xx:xx:xx 格式时间

```python
secs=12305

hour = secs//3600
minute = (secs - hour*3600)//60
second = (secs - hour*3600 - minute*60)

print('{0}:{1}:{2}'.format(hour,minute,second))
```
