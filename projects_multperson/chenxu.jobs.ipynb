```python
'''
@Description: 统一地图显示多个职业数据
@Author: Jkonel
@Date: 2020-06-23 20:45:42
@LastEditors: jkonel
@LastEditTime: 2020-06-24 10:56:42
'''
from pyecharts.charts import Map
from typing import List
import pyecharts.options as opts
from pyecharts.globals import ThemeType
from pyecharts.commons.utils import JsCode
from pyecharts.charts import Timeline, Grid, Bar, Map, Pie, Line


##原始数据结构例子

data =[{'job': 'cc++',
  'data': [{'province': '湖北', 'value': 13297.97},
   {'province': '陕西', 'value': 13919.4},
   {'province': '广东', 'value': 16545.04},
   {'province': '辽宁', 'value': 12418.89},
   {'province': '江西', 'value': 11456.12},
   {'province': '上海', 'value': 19506.49},
   {'province': '山西', 'value': 8800.0},
   {'province': '甘肃', 'value': 7375.0},
   {'province': '内蒙古', 'value': 13250.0},
   {'province': '河南', 'value': 9315.57},
   {'province': '广西', 'value': 8687.5},
   {'province': '北京', 'value': 18634.6},
   {'province': '云南', 'value': 20031.74},
   {'province': '河北', 'value': 15342.11},
   {'province': '四川', 'value': 14071.33},
   {'province': '吉林', 'value': 8200.0},
   {'province': '江苏', 'value': 15058.1},
   {'province': '天津', 'value': 11872.45},
   {'province': '重庆', 'value': 11600.56},
   {'province': '贵州', 'value': 9178.57},
   {'province': '新疆', 'value': 9000.0},
   {'province': '山东', 'value': 10739.11},
   {'province': '福建', 'value': 11938.52},
   {'province': '浙江', 'value': 15745.4},
   {'province': '黑龙江', 'value': 8057.69},
   {'province': '安徽', 'value': 13106.1},
   {'province': '湖南', 'value': 12255.24}]},
 {'job': 'embeded_hardware',
  'data': [{'province': '湖北', 'value': 10533.52},
   {'province': '陕西', 'value': 9738.06},
   {'province': '广东', 'value': 13342.9},
   {'province': '辽宁', 'value': 7075.0},
   {'province': '江西', 'value': 11214.29},
   {'province': '上海', 'value': 13602.67},
   {'province': '河南', 'value': 9315.46},
   {'province': '北京', 'value': 14298.61},
   {'province': '云南', 'value': 9000.0},
   {'province': '河北', 'value': 9800.0},
   {'province': '四川', 'value': 9308.14},
   {'province': '江苏', 'value': 11668.89},
   {'province': '天津', 'value': 9687.5},
   {'province': '重庆', 'value': 9250.0},
   {'province': '贵州', 'value': 7000.0},
   {'province': '山东', 'value': 9760.09},
   {'province': '福建', 'value': 11476.14},
   {'province': '浙江', 'value': 13074.66},
   {'province': '黑龙江', 'value': 6375.0},
   {'province': '安徽', 'value': 12121.77},
   {'province': '湖南', 'value': 11063.43}]},
 {'job': 'embeded_software',
  'data': [{'province': '湖北', 'value': 12163.08},
   {'province': '陕西', 'value': 11768.88},
   {'province': '广东', 'value': 14521.95},
   {'province': '辽宁', 'value': 8612.05},
   {'province': '宁夏', 'value': 7000.0},
   {'province': '江西', 'value': 10157.71},
   {'province': '上海', 'value': 16618.26},
   {'province': '甘肃', 'value': 14000.0},
   {'province': '内蒙古', 'value': 10416.5},
   {'province': '河南', 'value': 9675.16},
   {'province': '广西', 'value': 15423.08},
   {'province': '北京', 'value': 16314.26},
   {'province': '云南', 'value': 8383.2},
   {'province': '河北', 'value': 9924.23},
   {'province': '四川', 'value': 12578.61},
   {'province': '吉林', 'value': 8700.0},
   {'province': '江苏', 'value': 13037.19},
   {'province': '天津', 'value': 12161.44},
   {'province': '重庆', 'value': 10624.99},
   {'province': '山东', 'value': 10522.16},
   {'province': '福建', 'value': 11789.2},
   {'province': '浙江', 'value': 13657.7},
   {'province': '黑龙江', 'value': 8875.0},
   {'province': '安徽', 'value': 12721.63},
   {'province': '湖南', 'value': 11795.53}]},
 {'job': 'go',
  'data': [{'province': '湖北', 'value': 15552.06},
   {'province': '陕西', 'value': 13083.33},
   {'province': '广东', 'value': 18564.26},
   {'province': '辽宁', 'value': 7833.33},
   {'province': '上海', 'value': 19057.39},
   {'province': '河南', 'value': 13000.0},
   {'province': '广西', 'value': 12500.0},
   {'province': '北京', 'value': 25039.0},
   {'province': '云南', 'value': 10833.33},
   {'province': '河北', 'value': 17500.0},
   {'province': '四川', 'value': 13324.77},
   {'province': '江苏', 'value': 14727.08},
   {'province': '天津', 'value': 15000.0},
   {'province': '重庆', 'value': 11500.0},
   {'province': '山东', 'value': 10750.0},
   {'province': '福建', 'value': 13833.33},
   {'province': '浙江', 'value': 14393.47},
   {'province': '安徽', 'value': 12750.0},
   {'province': '湖南', 'value': 25000.0}]},
 {'job': 'java',
  'data': [{'province': '湖北', 'value': 10679.69},
   {'province': '陕西', 'value': 10931.75},
   {'province': '青海', 'value': 9555.56},
   {'province': '广东', 'value': 15214.97},
   {'province': '辽宁', 'value': 9469.22},
   {'province': '宁夏', 'value': 8813.46},
   {'province': '江西', 'value': 9585.85},
   {'province': '上海', 'value': 16883.39},
   {'province': '山西', 'value': 8847.83},
   {'province': '甘肃', 'value': 8889.62},
   {'province': '内蒙古', 'value': 9513.16},
   {'province': '河南', 'value': 9670.49},
   {'province': '广西', 'value': 7960.89},
   {'province': '北京', 'value': 15816.9},
   {'province': '云南', 'value': 9427.55},
   {'province': '海南', 'value': 9017.54},
   {'province': '河北', 'value': 7477.27},
   {'province': '四川', 'value': 10614.96},
   {'province': '西藏', 'value': 9625.0},
   {'province': '吉林', 'value': 7858.21},
   {'province': '江苏', 'value': 11830.07},
   {'province': '天津', 'value': 11051.19},
   {'province': '重庆', 'value': 10480.08},
   {'province': '贵州', 'value': 9415.12},
   {'province': '新疆', 'value': 9149.56},
   {'province': '山东', 'value': 8729.61},
   {'province': '福建', 'value': 32119.01},
   {'province': '浙江', 'value': 14083.25},
   {'province': '黑龙江', 'value': 7221.46},
   {'province': '安徽', 'value': 10313.34},
   {'province': '湖南', 'value': 10842.21}]},
 {'job': 'linux',
  'data': [{'province': '湖北', 'value': 10162.47},
   {'province': '陕西', 'value': 10542.43},
   {'province': '青海', 'value': 6022.73},
   {'province': '广东', 'value': 13384.59},
   {'province': '辽宁', 'value': 9946.14},
   {'province': '宁夏', 'value': 6863.64},
   {'province': '江西', 'value': 7364.56},
   {'province': '上海', 'value': 14628.8},
   {'province': '山西', 'value': 5557.69},
   {'province': '甘肃', 'value': 7031.25},
   {'province': '内蒙古', 'value': 7093.75},
   {'province': '河南', 'value': 8214.27},
   {'province': '广西', 'value': 6793.11},
   {'province': '北京', 'value': 14622.67},
   {'province': '云南', 'value': 7849.04},
   {'province': '海南', 'value': 9484.38},
   {'province': '河北', 'value': 6322.22},
   {'province': '四川', 'value': 11174.27},
   {'province': '西藏', 'value': 7041.67},
   {'province': '吉林', 'value': 6804.52},
   {'province': '江苏', 'value': 10589.11},
   {'province': '天津', 'value': 9512.81},
   {'province': '重庆', 'value': 9106.56},
   {'province': '贵州', 'value': 6513.88},
   {'province': '新疆', 'value': 7578.92},
   {'province': '山东', 'value': 8757.85},
   {'province': '福建', 'value': 9905.26},
   {'province': '浙江', 'value': 12333.97},
   {'province': '黑龙江', 'value': 7622.79},
   {'province': '安徽', 'value': 9235.34},
   {'province': '湖南', 'value': 9376.39}]},
 {'job': 'php',
  'data': [{'province': '湖北', 'value': 10116.54},
   {'province': '陕西', 'value': 10970.59},
   {'province': '青海', 'value': 7000.0},
   {'province': '广东', 'value': 12347.77},
   {'province': '辽宁', 'value': 8554.67},
   {'province': '江西', 'value': 7013.89},
   {'province': '上海', 'value': 15468.76},
   {'province': '山西', 'value': 6500.0},
   {'province': '甘肃', 'value': 9666.5},
   {'province': '河南', 'value': 9214.62},
   {'province': '广西', 'value': 7007.81},
   {'province': '北京', 'value': 16926.23},
   {'province': '云南', 'value': 7813.83},
   {'province': '海南', 'value': 8846.15},
   {'province': '河北', 'value': 10452.38},
   {'province': '四川', 'value': 10151.51},
   {'province': '吉林', 'value': 6263.64},
   {'province': '江苏', 'value': 10910.28},
   {'province': '天津', 'value': 11083.33},
   {'province': '重庆', 'value': 8797.17},
   {'province': '贵州', 'value': 8431.82},
   {'province': '山东', 'value': 8374.43},
   {'province': '福建', 'value': 10776.04},
   {'province': '浙江', 'value': 11954.53},
   {'province': '黑龙江', 'value': 6250.0},
   {'province': '安徽', 'value': 9591.65},
   {'province': '湖南', 'value': 11077.7}]},
 {'job': 'python',
  'data': [{'province': '湖北', 'value': 13157.81},
   {'province': '陕西', 'value': 10183.33},
   {'province': '广东', 'value': 16915.94},
   {'province': '辽宁', 'value': 8777.2},
   {'province': '宁夏', 'value': 8300.0},
   {'province': '江西', 'value': 10727.73},
   {'province': '上海', 'value': 18263.11},
   {'province': '甘肃', 'value': 5250.0},
   {'province': '内蒙古', 'value': 10500.0},
   {'province': '河南', 'value': 10505.68},
   {'province': '广西', 'value': 9000.0},
   {'province': '北京', 'value': 20037.92},
   {'province': '云南', 'value': 9178.02},
   {'province': '海南', 'value': 11000.0},
   {'province': '河北', 'value': 6027.78},
   {'province': '四川', 'value': 13886.68},
   {'province': '吉林', 'value': 12638.88},
   {'province': '江苏', 'value': 14474.38},
   {'province': '天津', 'value': 13373.0},
   {'province': '重庆', 'value': 10957.35},
   {'province': '贵州', 'value': 7190.43},
   {'province': '新疆', 'value': 10500.0},
   {'province': '山东', 'value': 9827.36},
   {'province': '福建', 'value': 12668.41},
   {'province': '浙江', 'value': 15847.43},
   {'province': '安徽', 'value': 11559.84},
   {'province': '湖南', 'value': 12122.96}]}]

'''
@description: 得到名字列表特定名字下标
@param : list,name
@return: index
'''
def get_list_index(list, str):
    index = 0
    for stri in list:
        if stri is str:
            return index
        else:
            index += 1
    return None

'''
@description: 
@param : job_name:'java';a:规范化数据;name:规范化名字列表
@return: grid_chart
'''
def get_chart(job_name, a, name):
    index = get_list_index(name, job_name)
    da = a[index]
    maxNum = max([x[1] for x in da])
    minNum = min([x[1] for x in da])

    map_chart = (
        Map()
        .add(
            series_name="",
            data_pair=da,
            zoom=1,
            center=[119.5, 34.5],
            is_map_symbol_show=False,
            itemstyle_opts={
                "normal": {"areaColor": "#323c48", "borderColor": "#404a59"},
                "emphasis": {
                    "label": {"show": Timeline},
                    "areaColor": "rgba(255,255,255, 0.5)",
                },
            },
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title = job_name,###
                subtitle="",
                pos_left="center",
                pos_top="top",
                title_textstyle_opts=opts.TextStyleOpts(
                    font_size=25, color="rgba(255,255,255, 0.9)"
                ),
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True,
                formatter=JsCode(
                    """function(params) {
                    if ('value' in params.data) {
                        return params.data.value[2] + ': ' + params.data.value[0];
                    }
                }"""
                ),
            ),
            visualmap_opts=opts.VisualMapOpts(
                is_calculable=True,
                dimension=0,
                pos_left="30",
                pos_top="center",
                range_text=["High", "Low"],
                range_color=["lightskyblue", "yellow", "orangered"],
                textstyle_opts=opts.TextStyleOpts(color="#ddd"),
                min_=10,
                max_=2000,
            ),
        )
    )
    bar_x_data = [x[0] for x in da]
    bar_y_data = [{"name": x[0], "value": x[1]} for x in da]
    bar = (
        Bar()
        .add_xaxis(xaxis_data=bar_x_data)
        .add_yaxis(
            series_name="",#job_name,##
            yaxis_data=bar_y_data,
            label_opts=opts.LabelOpts(
                is_show=True, position="right", formatter="{b} : {c}"
            ),
        )
        .reversal_axis()
        .set_global_opts(
            xaxis_opts=opts.AxisOpts(
                max_=maxNum, axislabel_opts=opts.LabelOpts(is_show=False)
            ),
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(is_show=False)),
            tooltip_opts=opts.TooltipOpts(is_show=False),
            visualmap_opts=opts.VisualMapOpts(
                is_calculable=True,
                dimension=0,
                pos_left="10",
                pos_top="top",
                range_text=["High", "Low"],
                range_color=["lightskyblue", "yellow", "orangered"],
                textstyle_opts=opts.TextStyleOpts(color="#ddd"),
                min_=minNum,
                max_=maxNum,
            ),
        )
    )

    pie = (
        Pie()
        .add(
            series_name = "",#job_name,##
            data_pair = da,###
            radius=["15%", "35%"],
            center=["80%", "82%"],
            itemstyle_opts=opts.ItemStyleOpts(
                border_width=1, border_color="rgba(0,0,0,0.3)"
            ),
        )
        .set_global_opts(
            tooltip_opts=opts.TooltipOpts(is_show=True, formatter="{b} {d}%"),
            legend_opts=opts.LegendOpts(is_show=False),
        )
    )
    
    grid_chart = (
        Grid()
            .add(
                bar,
                grid_opts=opts.GridOpts(
                    pos_left="10", pos_right="95%", pos_top="40%", pos_bottom="5"
                ),
            )
        .add(map_chart, grid_opts=opts.GridOpts())
        .add(pie, grid_opts=opts.GridOpts(pos_left="45%", pos_top="60%"))
    )
    return grid_chart


'''
@description: 地图主输出函数
@param : 规范原始数据
@return: 地图输出
'''
def multicolorful_map(data):
    #有多少name就有多少个a元素
    name = [x['job'] for x in data]
    a = [ [[x['province'],x['value']] for x in d['data'] ] for d in data]
    #a =[[x['province'],x['value']] for x in data]

    timeline = Timeline(
        init_opts=opts.InitOpts(width="960px", height="480px", theme=ThemeType.DARK)
    )
    indexs = 0
    for job_name in name:
        timeline.add(get_chart(job_name, a, name), time_point=indexs)
        indexs+=1

    timeline.add_schema(
        orient="vertical",
        is_auto_play=True,
        is_inverse=True,
        play_interval=5000,
        pos_left="null",
        pos_right="5",
        pos_top="20",
        pos_bottom="20",
        width="60",
        label_opts=opts.LabelOpts(is_show=True, color="#fff"),
    )
    return timeline

timelines = multicolorful_map(result_data_q1)
timelines.render_notebook()
```
