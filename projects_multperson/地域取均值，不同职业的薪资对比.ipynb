# python-all jobs

```python
import pandas as pd
import numpy as np
```

```python
python_all_jobs_new=pd.read_csv('date_2020-6-23/python_salary_norm_add_provence_city_area_del_none.csv')
python_all_jobs_new
```

```python
cols=['title','provence','salary_min','salary_max']

python_jobs=pd.DataFrame(python_all_jobs_new,columns=cols)
python_jobs
```

```python
# n1=python_jobs.groupby('provence').count()['title']
# print(n1)
n2=(python_jobs['salary_min'].mean()+python_jobs['salary_min'].mean())/2
print(n2)
```
